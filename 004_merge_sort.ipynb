```python
def mergesort(list):
    l = len(list)
    print("splitting list:", list)
    if l>1:
        m = l // 2
        left = list[:m]
        right = list[m:]
        mergesort(left)
        mergesort(right)
    
        print("====",left)
        print("+++",right)
        
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                list[k] = left[i]
                i+=1
            else:
                list[k] = right[j]
                j+=1
            k+=1
        
        while i < len(left):
            list[k] = left[i]
            i+=1
            k+=1
        
        while j < len(right):
            list[k] = right[j]
            j+=1
            k+=1
                    
        
```

```python
list = [4,5,2,7,1,9,3,5,8,1,9,3,0,11,88,33,77,55,11,00,33,66]
mergesort(list)
list
```
