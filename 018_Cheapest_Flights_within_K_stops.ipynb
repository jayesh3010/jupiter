```python
from typing import List
import sys
class Queue:
    def __init__(self ,items = [], count = 0):
        self.items = items
        self.count = count

    def enqueue(self ,data):
        self.items.insert(0 ,data)
        self.count += 1

    def dequeue(self):
        self.count -= 1
        return self.items.pop()

    def __str__(self) -> str:
        return(str(self.items) + "::" + str(self.count))

class Solution:
    def findCheapestPrice(self, n: int, flights: List[List[int]], src: int, dst: int, K: int) -> int:

        graph = dict()

        for s ,d ,p in flights:
            if s in graph:
                graph[s].append([d ,p])
            else:
                graph[s] = [[d ,p]]

        print("graph::",graph)

        queue = Queue()
        queue.enqueue([src,0])

        min_price = sys.maxsize
        print("min_price::",min_price)

        stops = 0

        while queue.count > 0:
            print("queue::", queue)

            size = len(queue.items)

            for _ in range(size):
                currentStop = queue.dequeue()
                print("currentStop::",currentStop)

                if currentStop[0] == dst :
                    min_price = min(min_price,currentStop[1])
                    print("min_price::", min_price)

                if currentStop[0] not in graph:
                    continue

                for nextStop in graph[currentStop[0]]:
                    print("nextStop::",nextStop)
                    if nextStop[1] + currentStop[1] > min_price :
                        continue
                    else:
                        queue.enqueue([nextStop[0],nextStop[1] + currentStop[1]])

            stops += 1
            if stops > K + 1 :
                break

        if min_price == sys.maxsize:
            min_price = -1

        return min_price

```

```python
from collections import deque,defaultdict
from typing import List
import  sys

class Solution:
    def findCheapestPrice(self, n: int, flights: List[List[int]], src: int, dst: int, K: int) -> int:

        dict = defaultdict(list)
        for s,d,p in flights:
            dict[s].append([d,p])

        print("dict::",dict)

        dq = deque()
        dq.append([src,0])

        minprice = sys.maxsize
        print("minprice::",minprice)

        numberofstops = 0

        while len(dq) > 0 :
            size = len(dq)
            print("dq::",dq)
            for _ in range(size) :

                stop = dq.popleft()

                if stop[0] == dst:
                    minprice = min(minprice, stop[1])
                    print("minprice::", minprice)
                    continue

                if stop[0] not in dict:
                    continue

                for node in dict[stop[0]]:
                    if node[1] + stop[1] > minprice:
                        continue   #pruning
                    dq.append([node[0],node[1] + stop[1]])

            numberofstops += 1
            if numberofstops > K+1 :
                break

        print("minprice::", minprice)
        return minprice if minprice < sys.maxsize else -1


s = Solution()
#print(s.findCheapestPrice(3,[[0,1,100],[1,2,100],[0,2,500]],0,2,0))
print(s.findCheapestPrice(4,[[0,1,1],[1,2,1],[2,3,1],[0,2,100]],0,3,1))

```
